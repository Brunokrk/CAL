```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
tempo = np.array([0.006, 0.065, 0.756, 8.103, 85.204, 362.550, 963.460])
n = np.array([30, 35, 40, 45, 50, 53, 55 ])

tempoB = np.array([0.003, 0.027, 0.264, 2.657])
nB = np.array([1000000, 10000000, 100000000, 1000000000])
```

```python
plt.plot(n, tempo)

```

```python
plt.plot(nB, tempoB)
```

```python
figure = plt.figure(figsize=(10,10), dpi=80)
sub_exponential = figure.add_subplot(211)
sub_linear = figure.add_subplot(212)

sub_exponential.set_title("Algoritmo Recursivo")
sub_exponential.set_xlabel("qtd elementos")
sub_exponential.set_ylabel("Segundos")
sub_linear.set_title("Algoritmo Não Recursivo")
sub_linear.set_xlabel("qtd elementos")
sub_linear.set_ylabel("Segundos")

sub_exponential.plot(n, tempo)
sub_linear.plot(nB, tempoB)
```
